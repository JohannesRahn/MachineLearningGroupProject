## **Recommendation algorithm based on KNN**

```python
# recommendation algorithm based on KNN 

import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler

# Assuming you have df_merged_users and df_merged_games_data DataFrames

# Create a user-item matrix
user_item_matrix = pd.pivot_table(df_merged_users, index='user_id', columns='app_id', values='hours', fill_value=0)

# Scale the user-item matrix to normalize the range of hours played
scaler = StandardScaler()
user_item_matrix_scaled = scaler.fit_transform(user_item_matrix)

# Create a sparse matrix for efficient computations
sparse_matrix = csr_matrix(user_item_matrix_scaled)

# Fit a KNN model on the sparse matrix
model_knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=10, n_jobs=-1)
model_knn.fit(sparse_matrix)

# Function to recommend a game to a user
def recommend_game(user_id, k=5):
    user_index = user_item_matrix.index.get_loc(user_id)
    distances, indices = model_knn.kneighbors(sparse_matrix[user_index], n_neighbors=k+1)
    
    # Get the indices of the nearest neighbors
    neighbor_indices = indices.flatten()[1:]
    
    # Get the games played by the neighbors
    neighbor_games = user_item_matrix.iloc[neighbor_indices].mean().sort_values(ascending=False)
    
    # Filter out the games that the user has already played
    user_games = set(user_item_matrix.iloc[user_index].loc[user_item_matrix.iloc[user_index] > 0].index)
    recommended_games = neighbor_games.loc[~neighbor_games.index.isin(user_games)].head(1)
    
    # Get the recommended game details
    recommended_game_details = df_merged_games_data.loc[df_merged_games_data['app_id'] == recommended_games.index[0]]
    
    return recommended_game_details

# Example usage: Recommend a game for user_id 12345
recommended_game = recommend_game(12345)
recommended_game
```

## **Recommendation algorithm using Decision Trees**

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Preprocessing and cleaning data
df_merged_games_data['rating'] = df_merged_games_data['rating'].fillna('Unknown')
df_merged_games_data['description'] = df_merged_games_data['description'].fillna('Unknown')

# Merge user and game data based on common app_id
df_merged = pd.merge(df_merged_users, df_merged_games_data, on='app_id', how='inner')

# Encode categorical features
label_encoder = LabelEncoder()
df_merged['rating_encoded'] = label_encoder.fit_transform(df_merged['rating'])
df_merged['is_recommended_encoded'] = df_merged['is_recommended'].astype(int)

# Define features and target variable
X = df_merged[['user_id', 'products', 'reviews', 'rating_encoded', 'positive_ratio', 'user_reviews', 'price_final', 'price_original', 'discount', 'steam_deck']]
y = df_merged['is_recommended_encoded']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a decision tree classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Recommend a game for a specific user (e.g., user_id = 12345)
user_id = 12345
user_data = df_merged[df_merged['user_id'] == user_id].drop_duplicates(subset='app_id')
user_data['predicted_recommendation'] = clf.predict(user_data[X.columns])

recommended_games = user_data[user_data['predicted_recommendation'] == 1]
print("Recommended games for user with user_id", user_id, "are:")
print(recommended_games[['title', 'rating']])
```

## **Recommendation system using random forest**

```python
# pipeline for recommenation system using random forest
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder

# Preprocessing and cleaning data
df_merged_games_data['rating'] = df_merged_games_data['rating'].fillna('Unknown')
df_merged_games_data['description'] = df_merged_games_data['description'].fillna('Unknown')

# Merge user and game data based on common app_id
df_merged = pd.merge(df_merged_users, df_merged_games_data, on='app_id', how='inner')
df_merged['is_recommended'] = df_merged['is_recommended'].astype(int)

# Define transformers for the pipeline
numeric_features = ['products', 'reviews', 'positive_ratio', 'user_reviews', 'price_final', 'price_original', 'discount']
numeric_transformer = StandardScaler()

categorical_features = ['rating', 'steam_deck']
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Define the pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])

# Define features and target variable
X = df_merged[['products', 'reviews', 'rating', 'positive_ratio', 'user_reviews', 'price_final', 'price_original', 'discount', 'steam_deck']]
y = df_merged['is_recommended']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the pipeline on the training data
pipeline.fit(X_train, y_train)

# Evaluate the pipeline on the testing data
score = pipeline.score(X_test, y_test)
print("Pipeline test score:", score)

# Recommend a game for a specific user (e.g., user_id = 12345)
user_id = 12345
user_data = df_merged[df_merged['user_id'] == user_id].drop_duplicates(subset='app_id')
user_data['predicted_recommendation'] = pipeline.predict(user_data[X.columns])

recommended_games = user_data[user_data['predicted_recommendation'] == 1]
print("Recommended games for user with user_id", user_id, "are:")
print(recommended_games[['title', 'rating']])
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder

# Preprocessing and cleaning data
df_merged_games_data['rating'] = df_merged_games_data['rating'].fillna('Unknown')
df_merged_games_data['description'] = df_merged_games_data['description'].fillna('Unknown')

# Merge user and game data based on common app_id
df_merged = pd.merge(df_merged_users, df_merged_games_data, on='app_id', how='inner')
df_merged['is_recommended'] = df_merged['is_recommended'].astype(int)

# Define transformers for the pipeline
numeric_features = ['products', 'reviews', 'positive_ratio', 'user_reviews', 'price_final', 'price_original', 'discount']
numeric_transformer = StandardScaler()

categorical_features = ['rating', 'steam_deck']
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Define the pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])

# Define features and target variable
X = df_merged[['products', 'reviews', 'rating', 'positive_ratio', 'user_reviews', 'price_final', 'price_original', 'discount', 'steam_deck']]
y = df_merged['is_recommended']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the parameter grid for GridSearchCV
param_grid = {
    'classifier__n_estimators': [10, 50, 100],
    'classifier__max_depth': [10, 50, 100]
}

# Perform GridSearchCV
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Evaluate the best model on the testing data
best_pipeline = grid_search.best_estimator_
score = best_pipeline.score(X_test, y_test)
print("Best pipeline test score:", score)

# Recommend a game for a specific user (e.g., user_id = 12345)
user_id = 12345
user_data = df_merged[df_merged['user_id'] == user_id].drop_duplicates(subset='app_id')
user_data['predicted_recommendation'] = best_pipeline.predict(user_data[X.columns])

recommended_games = user_data[user_data['predicted_recommendation'] == 1]
print("Recommended games for user with user_id", user_id, "are:")
print(recommended_games[['title', 'rating']])
```
